# Dataset EDA


```python
import os
import random
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
sns.set_style("whitegrid")
```

```python
TRAIN_DIR = "../data"
```

## Basic information


```python
image_files = [
    f for f in os.listdir(TRAIN_DIR) if f.lower().endswith((".png", ".jpg", ".jpeg"))
]

print(f"Total images: {len(image_files)}")
```

```python
extensions = [os.path.splitext(f)[1].lower() for f in image_files]
print("File type distribution:", Counter(extensions))
```

## Image Dimensions


```python
widths, heights = [], []

for img_name in image_files[:5000]:
    try:
        with Image.open(os.path.join(TRAIN_DIR, img_name)) as img:
            widths.append(img.width)
            heights.append(img.height)
    except:
        continue


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(widths, bins=20)
plt.title("Width distribution")

plt.subplot(1, 2, 2)
plt.hist(heights, bins=20)
plt.title("Height distribution")

plt.tight_layout()
plt.show()
```

## Sample images preview


```python
def plot_sample_images(image_dir, image_files, n=25):
    plt.figure(figsize=(10, 10))

    for i in range(n):
        img_path = os.path.join(image_dir, random.choice(image_files))
        img = Image.open(img_path)
        plt.subplot(5, 5, i + 1)
        plt.imshow(img)
        plt.axis("off")
    plt.tight_layout()
    plt.show()


plot_sample_images(TRAIN_DIR, image_files)
```
